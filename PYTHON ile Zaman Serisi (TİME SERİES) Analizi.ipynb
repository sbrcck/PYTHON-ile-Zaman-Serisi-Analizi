# PYTHON ile Zaman Serisi (TİME SERİES) Analizi

Zaman serisi analizi bir zaman serisi veri setindeki kalıpları analiz etmek ve bulmak anlamına gelir. Bir zaman serisi veri seti bir zaman aralığında toplanan bir veri dizisidir. Hisse senedi fiyatları, aylık satış verileri, günlük satış verileri, saatlik web sitesi trafiği vb. veriler zaman serisi vereilerine örnektir. 

### ZAMAN SERİSİ ANALİZİ 

Bir zaman serisinin zaman aralığı yıllık, aylık, haftalık, günlük ve hatta saatlik olabilir ama coğu problemlerde verilerinizi analiz etme yöntemleriniz aynı olacaktır.

### PYTHON ile Zaman Serisi (TİME SERİES) Analizi
Gerekli python kütüphanelerini ve bir zaman serisi verisi setini içe aktararak analizimize başlıyoruz.


```python
import pandas as pd
import yfinance as yf
import datetime
from datetime import date, timedelta
today = date.today()

d1 = today.strftime("%Y-%m-%d")
end_date = d1
d2 = date.today() - timedelta(days=720)
d2 = d2.strftime("%Y-%m-%d")
start_date = d2

data = yf.download('AAPL', 
                      start=start_date, 
                      end=end_date, 
                      progress=False)
print(data.head())
```

Yukarıdaki kodda, en son hisse senedi fiyat verilerini çıkarmak için yfinance API'sini kullandım.Şimdi Apple'ın hisse senedi fiyatlarındaki eğilimleri görmek için bir çizgi grafiğini görselleştirelim

```python
import plotly.express as px
figure = px.line(data, x = data.index, 
                 y = "Close", 
                 title = "Time Series Analysis (Line Plot)")
figure.show()
```

Çizgi grafiği, Zaman serisi analizi üzerinde çalışırken en iyi görselleştirme araçlarından biridir. Yukarıdaki kodda Apple'ın kapanış fiyatlarındaki trendleri görselleştiriyorum. İmleci satıra yerleştirirseniz, imlecinizin bulunduğu veri noktasının tam tarihinde Kapanış fiyatını göreceksiniz.

Şimdi Apple'ın açılış, en yüksek, en düşük ve kapanış fiyatlarındaki eğilimleri görmek için bir mum grafiğini görselleştirelim

```python
import plotly.graph_objects as go
figure = go.Figure(data=[go.Candlestick(x = data.index,
                                        open = data["Open"], 
                                        high = data["High"],
                                        low = data["Low"], 
                                        close = data["Close"])])
figure.update_layout(title = "Time Series Analysis (Candlestick Chart)", 
                     xaxis_rangeslider_visible = False)
figure.show()
```

Bir finansal enstrümanın zaman serisi analizinde bir mum grafiği her zaman yardımcı olur. İmleci yukarıdaki mum grafiğinde herhangi bir noktaya yerleştirirseniz, imlecinizin bulunduğu tarihte Apple'ın tüm fiyatlarını (açılış, yüksek, düşük ve kapanış) göreceksiniz. Bu grafiğin kırmızı çizgileri fiyatların düştüğünü, yeşil çizgileri ise fiyatların arttığını gösteriyor.

Şimdi dönem boyunca kapanış fiyatlarının eğilimlerini görselleştirmek için bir çubuk grafiğini görselleştirelim

```python
figure = px.bar(data, x = data.index, 
                y = "Close", 
                title = "Time Series Analysis (Bar Plot)" )
figure.show()
```

Yukarıdaki çubuk grafiği, uzun vadeli senaryoda hisse senedi fiyatlarındaki artışı göstermektedir. Çizgi grafik ve mum grafiği, fiyatın artış ve düşüşünü gösterir, ancak uzun vadede fiyat artış ve düşüşünü görmek istiyorsanız, her zaman bir çubuk grafiği tercih etmelisiniz.

İki belirli tarih arasındaki hisse senedi fiyatlarını analiz etmek istiyorsanız, bunu nasıl yapabileceğiniz aşağıda açıklanmıştır

```python
figure = px.line(data, x = data.index, 
                 y = 'Close', 
                 range_x = ['2021-07-01','2021-12-31'], 
                 title = "Time Series Analysis (Custom Date Range)")
figure.show()
```

Bir zaman serisi verilerini analiz etmenin en iyi yollarından biri, çıktı görselleştirmesinin kendisinde zaman aralığını manuel olarak seçebileceğiniz etkileşimli bir görselleştirme oluşturmaktır. Bunu yapmanın bir yolu, görselleştirmenizin altına bir kaydırıcı ve görselleştirmenizin üzerindeki zaman aralıklarını kontrol etmek için düğmeler eklemektir. 

Çıktının kendisinde zaman aralıklarını seçebileceğiniz etkileşimli bir mum grafiğini nasıl oluşturabileceğiniz aşağıda açıklanmıştır

```python
figure = go.Figure(data = [go.Candlestick(x = data.index,
                                        open = data["Open"], 
                                        high = data["High"],
                                        low = data["Low"], 
                                        close = data["Close"])])
figure.update_layout(title = "Time Series Analysis (Candlestick Chart with Buttons and Slider)")

figure.update_xaxes(
    rangeslider_visible = True,
    rangeselector = dict(
        buttons = list([
            dict(count = 1, label = "1m", step = "month", stepmode = "backward"),
            dict(count = 6, label = "6m", step = "month", stepmode = "backward"),
            dict(count = 1, label = "YTD", step = "year", stepmode = "todate"),
            dict(count = 1, label = "1y", step = "year", stepmode = "backward"),
            dict(step = "all")
        ])
    )
)
figure.show()
```
